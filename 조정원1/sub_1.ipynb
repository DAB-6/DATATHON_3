```python
import pandas as pd

# 1. 파일 로드
try:
    df = pd.read_csv('veterinary_clinical_data.csv')
except FileNotFoundError:
    print("오류: 'veterinary_clinical_data.csv' 파일을 찾을 수 없습니다. 파일 경로를 확인해 주세요.")
    exit()

# 2. 동물 종류(AnimalName) 및 품종(Breed)별 의료 기록(MedicalHistory) 빈도수 계산
# 두 컬럼을 기준으로 그룹화하고, MedicalHistory에 대해 value_counts()를 적용합니다.
# unstack(fill_value=0)을 사용하여 결과를 표 형태로 변환합니다.
analysis_result = df.groupby(['AnimalName', 'Breed'])['MedicalHistory'].value_counts().unstack(fill_value=0)

# 3. 결과 출력
print("=======================================================================")
print("      동물 종류(AnimalName) 및 품종(Breed)별 의료 기록 빈도 분석")
print("=======================================================================")
# 결과를 보기 쉽게 출력합니다.
print(analysis_result)
print("-----------------------------------------------------------------------")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Matplotlib에서 한글 폰트 설정 (필요시 주석 해제하여 사용)
# import matplotlib.font_manager as fm
# import platform
# if platform.system() == 'Windows':
#     plt.rcParams['font.family'] = 'Malgun Gothic'
# else: # Mac or Linux
#     plt.rcParams['font.family'] = 'DejaVu Sans' # 기본 폰트 설정
# plt.rcParams['axes.unicode_minus'] = False # 마이너스 폰트 깨짐 방지

# 1. 파일 로드
try:
    df = pd.read_csv('veterinary_clinical_data.csv')
except FileNotFoundError:
    print("오류: 'veterinary_clinical_data.csv' 파일을 찾을 수 없습니다. 파일 경로를 확인해 주세요.")
    exit()

# 2. 데이터 분리 및 준비
df_dog = df[df['AnimalName'] == 'Dog'].copy()
df_cat = df[df['AnimalName'] == 'cat'].copy()

# 모든 고유한 의료 기록 항목 리스트 추출
medical_history_types = df['MedicalHistory'].unique()
# MedicalHistory 항목이 너무 많으면 시각화가 복잡해지므로, 최대 6~8개 정도만 선택할 수 있습니다.
# 여기서는 전체를 사용하되, 개별 그래프를 생성합니다.

# ---------------------------------- 3. 개 (Dog) 그룹 시각화 ----------------------------------

print("=======================================================================")
print("                   🐶 개 품종별 의료 기록 항목별 빈도 분석")
print("=======================================================================")

# 서브플롯 레이아웃 계산 (한 줄에 최대 3개 그래프)
n_plots_dog = len(medical_history_types)
n_cols = 3
n_rows = int(np.ceil(n_plots_dog / n_cols))

fig_dog, axes_dog = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(18, 5 * n_rows))
axes_dog = axes_dog.flatten() # 1차원 배열로 평탄화

dog_breeds_order = df_dog['Breed'].value_counts().index # X축 정렬 기준

for i, history_type in enumerate(medical_history_types):
    if i < len(axes_dog): # 서브플롯 개수 내에서만 반복
        # 특정 의료 기록 항목만 필터링
        df_filtered = df_dog[df_dog['MedicalHistory'] == history_type].copy()
        
        # 품종별 빈도 계산
        breed_counts = df_filtered['Breed'].value_counts().reset_index()
        breed_counts.columns = ['Breed', 'Count']
        
        # 막대 그래프 생성
        sns.barplot(
            data=breed_counts,
            x='Breed',
            y='Count',
            order=dog_breeds_order, # 모든 그래프의 X축 순서를 동일하게 유지
            ax=axes_dog[i],
            palette='viridis' 
        )
        
        axes_dog[i].set_title(f'개: "{history_type}" 빈도', fontsize=12)
        axes_dog[i].set_xlabel('품종', fontsize=10)
        axes_dog[i].set_ylabel('빈도 (Count)', fontsize=10)
        plt.setp(axes_dog[i].get_xticklabels(), rotation=45, ha='right')

# 사용하지 않는 서브플롯 숨기기
for j in range(i + 1, len(axes_dog)):
    fig_dog.delaxes(axes_dog[j])

fig_dog.suptitle('개 품종별 의료 기록 항목별 상세 빈도', fontsize=20, y=1.02)
plt.tight_layout()
plt.show()


# ---------------------------------- 4. 고양이 (Cat) 그룹 시각화 ----------------------------------

print("\n=======================================================================")
print("                   🐈 고양이 품종별 의료 기록 항목별 빈도 분석")
print("=======================================================================")

# 서브플롯 레이아웃 계산
n_plots_cat = len(medical_history_types)
n_rows = int(np.ceil(n_plots_cat / n_cols))

fig_cat, axes_cat = plt.subplots(nrows=n_rows, ncols=n_cols, figsize=(18, 5 * n_rows))
axes_cat = axes_cat.flatten() # 1차원 배열로 평탄화

cat_breeds_order = df_cat['Breed'].value_counts().index # X축 정렬 기준

for i, history_type in enumerate(medical_history_types):
    if i < len(axes_cat): # 서브플롯 개수 내에서만 반복
        # 특정 의료 기록 항목만 필터링
        df_filtered = df_cat[df_cat['MedicalHistory'] == history_type].copy()
        
        # 품종별 빈도 계산
        breed_counts = df_filtered['Breed'].value_counts().reset_index()
        breed_counts.columns = ['Breed', 'Count']

        # 막대 그래프 생성
        sns.barplot(
            data=breed_counts,
            x='Breed',
            y='Count',
            order=cat_breeds_order, # 모든 그래프의 X축 순서를 동일하게 유지
            ax=axes_cat[i],
            palette='magma'
        )
        
        axes_cat[i].set_title(f'고양이: "{history_type}" 빈도', fontsize=12)
        axes_cat[i].set_xlabel('품종', fontsize=10)
        axes_cat[i].set_ylabel('빈도 (Count)', fontsize=10)
        plt.setp(axes_cat[i].get_xticklabels(), rotation=45, ha='right')

# 사용하지 않는 서브플롯 숨기기
for j in range(i + 1, len(axes_cat)):
    fig_cat.delaxes(axes_cat[j])

fig_cat.suptitle('고양이 품종별 의료 기록 항목별 상세 빈도', fontsize=20, y=1.02)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. 파일 로드
try:
    df = pd.read_csv('veterinary_clinical_data.csv')
except FileNotFoundError:
    print("오류: 'veterinary_clinical_data.csv' 파일을 찾을 수 없습니다. 파일 경로를 확인해 주세요.")
    exit()

# 2. '개(Dog)' 데이터만 필터링
df_dog = df[df['AnimalName'] == 'Dog'].copy()

# 3. 모든 고유한 의료 기록 항목 리스트 추출
medical_history_types = df_dog['MedicalHistory'].unique()

# 4. 각 의료 기록 항목별 개 품종 순위를 계산하고 저장
dog_rankings = []
for history in medical_history_types:
    # 1. 현재 의료 기록 항목만 필터링
    df_filtered = df_dog[df_dog['MedicalHistory'] == history].copy()
    
    # 2. 품종별 빈도를 계산하고 (빈도순으로 자동 정렬)
    breed_counts = df_filtered['Breed'].value_counts().reset_index()
    breed_counts.columns = ['품종 (Breed)', '빈도 (Count)']
    
    # 3. 순위 컬럼 추가
    breed_counts['순위 (Rank)'] = breed_counts.index + 1
    
    # 4. 결과 저장
    dog_rankings.append({
        'HistoryType': history,
        'Ranking': breed_counts[['순위 (Rank)', '품종 (Breed)', '빈도 (Count)']]
    })

# 5. Markdown 보고서 생성
markdown_content = "### 🐕 개(Dog) 품종별 의료 기록 빈도 순위 (내림차순)\n\n"
markdown_content += "이 표는 **개 품종**에 한정하여, 각 의료 기록 항목을 가장 많이 가진 품종 순서대로 정리한 결과입니다. \n\n"
markdown_content += "---"

for result in dog_rankings:
    history = result['HistoryType']
    ranking_df = result['Ranking']
    
    # 품종이 하나도 없는 경우 (이론상 dog_rankings 리스트에 들어가지 않으나, 혹시 모를 경우 대비)
    if ranking_df.empty:
        continue
        
    markdown_content += f"\n\n## 🏆 {history} 순위\n\n"
    markdown_content += ranking_df.to_markdown(index=False)
    markdown_content += "\n\n---"

# 6. 결과 출력 (Markdown 파일로 생성)
print(markdown_content)
```

```python
#개의 의료기록 하나당 대표증상들
import pandas as pd
import numpy as np

df = pd.read_csv('veterinary_clinical_data.csv')
conditions = [
    (df['Age'] <= 0.5),             # .5년 이하
    (df['Age'] > 0.5) & (df['Age'] <= 0.75), # .5년 초과 1년 이하
    (df['Age'] > 0.75) & (df['Age'] <= 8), # 1년 초과 6년 이하
    (df['Age'] > 8)              # 6년 초과
]


choices = ['weaning', 'half_mature','mature', 'senior']

df['growth_stage'] = np.select(conditions, choices, default='미분류')
```

```python
df_Golden_Retriever=df[df['Breed']=='Yorkshire Terrier']
print(f"Yorkshire Terrier 품종의 총 데이터 수: {len(df_Golden_Retriever)}")
print(f"{df_Golden_Retriever['MedicalHistory'].value_counts()}")
```

```python
df_Golden_Retriever_weaning=df_Golden_Retriever[df_Golden_Retriever['growth_stage']=='weaning']
print(f"Yorkshire Terrier 품종 weaning의 총 데이터 수: {len(df_Golden_Retriever_weaning)}")
print(f"{df_Golden_Retriever_weaning['MedicalHistory'].value_counts()}")
```

```python
df_Golden_Retriever_half=df_Golden_Retriever[df_Golden_Retriever['growth_stage']=='half_mature']
print(f"Yorkshire Terrier 품종 half_mature의 총 데이터 수: {len(df_Golden_Retriever_half)}")
print(f"{df_Golden_Retriever_half['MedicalHistory'].value_counts()}")
```

```python
df_Golden_Retriever_mature=df_Golden_Retriever[df_Golden_Retriever['growth_stage']=='mature']
print(f"Yorkshire Terrier 품종 mature의 총 데이터 수: {len(df_Golden_Retriever_mature)}")
print(f"{df_Golden_Retriever_mature['MedicalHistory'].value_counts()}")
```

```python
df_Golden_Retriever_senior=df_Golden_Retriever[df_Golden_Retriever['growth_stage']=='senior']
print(f"Yorkshire Terrier 품종 seinor의 총 데이터 수: {len(df_Golden_Retriever_senior)}")
print(f"{df_Golden_Retriever_senior['MedicalHistory'].value_counts()}")
```
