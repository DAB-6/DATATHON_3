```python
# 1. 'digestion' 컬럼 생성 (pet_health_issue_dig)
df_dropped_dry['pet_health_issue_dig'] = (
    df_dropped_dry['pet_health_issue_list']
    .astype(str) # 문자열로 강제 변환하여 NaN 문제 등을 방지
    .str.split(', ') # 'digestion, joints'를 ['digestion', 'joints'] 리스트로 분리
    .apply(lambda x: 1 if 'digestion' in x else 0)
)

# 2. 'joints' 컬럼 생성 (pet_health_issue_joint)
df_dropped_dry['pet_health_issue_joint'] = (
    df_dropped_dry['pet_health_issue_list']
    .astype(str)
    .str.split(', ')
    .apply(lambda x: 1 if 'joints' in x else 0)
)

# 3. 'skin and coat' 컬럼 생성 (pet_health_issue_skin)
df_dropped_dry['pet_health_issue_skin'] = (
    df_dropped_dry['pet_health_issue_list']
    .astype(str)
    .str.split(', ')
    .apply(lambda x: 1 if 'skin and coat' in x else 0)
)

print("\n--- [수정 완료] 새로운 컬럼이 추가된 데이터프레임의 상위 5개 행 ---")
# 새로 생성된 컬럼을 포함하여 몇 개 행을 다시 확인
print(df_dropped_dry[['pet_health_issue_list', 
                      'pet_health_issue_dig', 
                      'pet_health_issue_joint', 
                      'pet_health_issue_skin']].head())
```

```python
allergen_list = [
    'beef', 'chicken', 'dairy', 'egg', 'lamb', 'soya', 'wheat',
    'fish', 'grain', 'maize'
]

df_split = (
    df_dropped_dry['pet_allergen_list']
    .astype(str)
    .str.lower() 
    .str.split()  
    .apply(lambda x: [item.strip() for item in x]) # 분리된 각 요소의 앞뒤 공백 제거
)


for allergen in allergen_list:
    new_col_name = f'pet_allergen_{allergen}'
    
    df_dropped_dry[new_col_name] = df_split.apply(lambda x: 1 if allergen in x else 0)

df_dropped_dry
```

```python
risk_data = {
    ('True', 'female'):  {'joint': 0.1017, 'skin': 0.4704, 'dig': 0.0312, 'inner': 0.3880}, # 중성화 여
    ('False', 'female'): {'joint': 0.0634, 'skin': 0.3065, 'dig': 0.012, 'inner': 0.2862}, # 중성화 안 한 여
    ('True', 'male'):    {'joint': 0.118, 'skin': 0.537, 'dig': 0.039, 'inner': 0.280}, # 중성화 남
    ('False', 'male'):   {'joint': 0.1228, 'skin': 0.505, 'dig': 0.03, 'inner': 0.319}, # 중성화 안 한 남
}

# 1. 'neutered' 컬럼이 True/False 불리언이 아닌 문자열('True', 'False')일 수 있으므로 문자열로 통일합니다.
#    'gender' 컬럼도 혹시 모를 문제 방지를 위해 문자열로 통일하고 공백을 제거합니다.
df_dropped_dry['neutered_str'] = df_dropped_dry['neutered'].astype(str).str.lower().str.strip()
df_dropped_dry['gender_str'] = df_dropped_dry['gender'].astype(str).str.lower().str.strip()

# 2. 새로운 '발병률 위험도' 컬럼 생성 및 값 할당
#    모든 새로운 컬럼을 0.0으로 초기화합니다.
health_issues = ['joint', 'skin', 'dig', 'inner']
for issue in health_issues:
    new_col_name = f'neutered_gender_{issue}'
    df_dropped_dry[new_col_name] = 0.0

# 3. 모든 그룹 조건에 따라 발병률 값을 할당
for (neutered_val, gender_val), risks in risk_data.items():
    # 현재 그룹에 해당하는 행을 찾는 조건
    condition = (df_dropped_dry['neutered_str'] == neutered_val.lower()) & \
                (df_dropped_dry['gender_str'] == gender_val.lower())
    
    # 각 건강 이슈별로 해당 발병률 값을 대입
    for issue in health_issues:
        new_col_name = f'neutered_gender_{issue}'
        risk = risks[issue]
        
        # 조건에 맞는 행의 새 컬럼에 해당 발병률 값을 할당
        df_dropped_dry.loc[condition, new_col_name] = risk

print("\n--- [모든 그룹의 발병률 컬럼 생성 완료] 결과 확인 ---")
# 중성화된 암컷과 중성화 안 한 수컷의 데이터 5개 행을 비교하여 확인
print(df_dropped_dry[['neutered_str', 'gender_str', 
                      'neutered_gender_joint', 
                      'neutered_gender_skin']].head())

# 임시 컬럼 제거
df_dropped_dry = df_dropped_dry.drop(columns=['neutered_str', 'gender_str'])

df
```
